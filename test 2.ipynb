```python
import pandas as pd
from sklearn.cluster import KMeans
```

```python
dataset=pd.read_csv('Oportunities.csv')
dataset['Monto']=dataset['Monto'].fillna(dataset['Monto'].mean())
pd.isnull(dataset['Monto']).values.ravel().sum() 
```

```python
#dataset=dataset.sort_values(by='Monto',ascending=True)
#dataset = dataset.reset_index(drop=True)

model_m=KMeans(n_clusters=5).fit(dataset[['Monto']])
clust_m=pd.Series(model_m.labels_)
dataset['p_m']=clust_m
```

```python
dataset.head(20)
```

```python
limits={}
lim_clusts=[2,4,7]

for index,i in enumerate(lim_clusts):
    key = 'limit'+str(index+1)
    value = i
    limits[key] = value

print(limits)    
    
for index,j in enumerate(limits):
    print(j+': '+str(limits[j])+' -->'+str(index))
    if j=='limit1':
        dataset.loc[(dataset['Tickets'] <= limits[j]), 'p_m'] = limits[j]
        temp=limits[j]
    else:
        dataset.loc[(dataset['Tickets'] <= limits[j]) & (dataset['Tickets'] > temp), 'p_m'] = limits[j]
        temp=limits[j]
 
           
dataset.head(20)
```
